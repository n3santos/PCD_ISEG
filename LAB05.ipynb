 4. Convert the string into a list.

```python
listSpeech = speech.split()
print (listSpeech)
```

5. Use a dictionary to get the list of words and their frequencies.

```python
from collections import Counter
counts = Counter(listSpeech)
print(counts)
```

```python
#Opção para estar em dicionário
dictSpeech = dict(counts)
print(dictSpeech)
```

```python
ListOfWords = {}
for word in listSpeech:
    if word in ListOfWords:
        ListOfWords[word] += 1
    else:
        ListOfWords [word]= 1
print (ListOfWords)
```

6. Sort the dictionary alphabetically.

```python
dictionaryitems = dictSpeech.items()
sorteditems = sorted (dictionaryitems)
print (sorteditems)
```

```python
alphabet = []
for key, value in dictSpeech.items():
    alphabet.append((value, key))

alphabet.sort()
print (alphabet)
```
